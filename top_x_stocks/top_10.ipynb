```python
import pandas as pd
# import matplotlib.pyplot as plt
import numpy as np
import os
import yaml
```

```python
result_list = os.listdir('../regressor_results_partial_0-6/')
top_10 = pd.DataFrame(columns=['stock', 'buy_threshold', 'sell_threshold', 'lookback_period', 'total_transactions', 'adjusted_P&L'])

for stock in result_list:
    df = pd.read_csv(f'../regressor_results_partial_0-6/{stock}')
    if df.shape[0] > 0:
        top_10.loc[len(df)] = [
                        stock.split('_')[0],
                        df.loc[0, 'buy_threshold'],
                        df.loc[0, 'sell_threshold'],
                        df.loc[0, 'lookback_period'],
                        df.loc[0, 'total_transactions'],
                        int(df.loc[0, 'adjusted_P&L'])
                        ]

top_10 = top_10.sort_values(by=['adjusted_P&L'], ascending=False)
top_10 = top_10.iloc[:10, :]
top_10.reset_index(inplace=True, drop=True)
top_10.iloc[:, 0:4].to_pickle('./top_10.pkl')
```

```python
top_10_dict = {value: { 'symbol': value } for value in top_10['stock']}

with open('../sample.yaml', 'r') as yaml_file:
    new_yaml = yaml.load(yaml_file, Loader=yaml.SafeLoader)
    new_yaml.update(top_10_dict)
    
with open('../sample.yaml', 'w') as yaml_file:
    yaml.safe_dump(new_yaml, yaml_file, sort_keys=False)
```
